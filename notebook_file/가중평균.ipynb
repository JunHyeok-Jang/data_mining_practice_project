```python
import pandas as pd
```

```python
df = pd.read_csv('프로기록있는선수들통과.csv')
```

```python
df = df.reset_index(drop=True)
```

```python
k = df.groupby(['선수명','대학교','대회년도']).sum()
```

```python
k = k[k.columns[1:]]
```

```python
a = k.groupby(['선수명','대학교']).count()
```

```python
k_new = k.merge(a, how='left',on = ['선수명','대학교'])
```

```python
k_new = k_new[k_new.columns[:29]]
k_new.columns = ['경기수', '득점 전체', '득점 평균', '2점슛 성공', '2점슛 시도', '3점슛 성공', '3점슛 시도',
       '자유투 성공', '자유투 시도', '공격 리바운드', '수비 리바운드', '평균 리바운드', '전체 어시스트',
       '평균 어시스트', '스틸', '블록', '덩크슛 성공', '덩크슛 시도', '파울(자유튜유)', '파울(자유투무)',
       'IDF', '테크니컬 파울', '굿디펜스', '속공', 'FBF', '턴오버', '팀 리바운드', '팀 턴오버',
       '년수']
k_new
```

```python
emptydf = pd.DataFrame([])

#수정가능 - 마지막년도 가중치 ~ 첫해 가중치 순서임
year2 = [0.7,0.3]
year3 = [0.5,0.3,0.2]
year4 = [0.4,0.3,0.2,0.1]
year5 = [0.4,0.25,0.2,0.1,0.05]

#선수를 구분하기 위한 counter, 선수의 case(년수)를 구분하기 위한 total
counter = 0
total = 0

for idx,row in k_new.iterrows():
    if counter == 0:
        total = int(row['년수'])
        counter = int(row['년수'])    
    if total == 1:
        emptydf = emptydf.append(row,ignore_index = False)
    elif total == 2:
        row= year2[counter-1]*row
        emptydf = emptydf.append(row,ignore_index = False)
    elif total == 3:
        row = year3[counter-1]*row
        emptydf = emptydf.append(row,ignore_index = False)
    elif total == 4:
        row = year4[counter-1]*row
        emptydf = emptydf.append(row,ignore_index = False)
    elif total == 5:
        row = year5[counter-1]*row
        emptydf = emptydf.append(row,ignore_index = False)
    counter -=1
```

```python
result = emptydf.set_index(k_new.index).reset_index()
result
```

```python
realresult = result.groupby(['선수명','대학교']).sum().reset_index()
realresult
```

```python
realresult.to_csv('가중평균.csv',encoding = 'utf-8-sig')
```
